## Polynomial line 

```
import numpy as np
import matplotlib.pyplot as plt


# Create polyline object  with degree 2 
p1 = np.poly1d([1, 2, 3])
print(p1)
print(p1.roots)
x = np.linspace(-10,10,200)
plt.plot(x,p1(x))

```

## Fitting poly line จากข้อมูล

```
x = [-10 , -5 , 0, 5, 10 , 15 , 20  , 25 ,30 ]

y = [1.341, 1.316 , 1.293  , 1.6  , 1.247  , 1.225  , 1.204 , 1.184   , 1.164]

plt.plot(x,y,'ro')
```

```
# Fitting polyline degree 3 ลงบนข้อมูล
deg = 10



coeff = np.polyfit(x, y, deg)
p = np.poly1d(coeff)


print( p) # prints the polynomial expression


y_fitted = p(x) # computes the polynomial at the x points

# Use red circles for data points and a blue line for the polyn.
plt.plot(x,y,'ro', x , y_fitted)
```

```
# Create polyline object  
p1 = np.poly1d([-2, 3])
print(p1)

p2 = np.poly1d([2, -350])
print(p2)


Nsampling = 30
x1 = np.linspace(-100,100,Nsampling)
y1 = p1(x1) + 100*np.sin(x1) + 15*np.random.randn(Nsampling)

x2 = np.linspace(100,300, Nsampling)
y2 = p2(x2) + 100*np.cos(x2) + 15*np.random.randn(Nsampling)

x = np.append(x1 , x2 )
y = np.append(y1 , y2 )
plt.plot(x,y,'ro')
```
