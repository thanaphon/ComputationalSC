## Definition: 

Histrogram  https://en.wikipedia.org/wiki/Histogram

ฮิสโตแกรม (Histogram) คือกราฟแท่งแบบเฉพาะที่แสดงความสัมพันธ์ระหว่างข้อมูลเป็นหมวดหมู่ที่เรียกว่าชั้นข้อมูลกับความถี่ของข้อมูล เพื่อดูการกระจายของข้อมูล ลักษณะของข้อมูลที่เป็นหมวดหมู่จะเรียงลำดับจากน้อยไปหามากโดยจำนวนหมวดหมู่ของข้อมูลจะจัดตามความเหมาะสม โดยแกนตั้งจะเป็นตัวเลขแสดง “ ความถี่ ” และแกนนอนจะเป็นข้อมูลคุณสมบัติของสิ่งที่เราสนใจ แท่งกราฟแต่ละแท่งจะมีความกว้างเท่ากันซึ่งเท่ากับกว้างของชั้นข้อมูล ส่วนความสูงของกราฟแต่ละแท่งนั้นจะสูงเท่ากับจำนวนความถี่ของแต่ละชั้นข้อมูล

## ตัวอย่าง โค้ด Python ที่ใช้ในการนับความถี่ข้อมูลโดยอาศัย Data Dict 

```
# Need not be sorted, necessarily
# ตัวอย่าง input array  


a = (0, 1, 1, 1, 2, 3, 7, 7, 23)

# สร้างฟังก์ชั่นเพื่อช่วยในการนับ โดยอาศัยการ Hashing 
def count_elements(seq) -> dict: # ระบุ Type ของการ return (มีหรือไม่ก็ได้) 
    ''' Tally elements from "sequence" ''' 
    hist = {}  # สร้าง Data dict ชื่อ hist 
    for i in seq:
        hist[i] = hist.get(i, 0) + 1   # hist.get('key' , default = 0 )  ค่า default เป็นค่าที่ return ออกมากรณีที่ ยังไม่มี ค่า key value ใน data dict
    return hist

counted = count_elements(a)
counted
```

## HOW to read image from URL website

```
from PIL import Image  # ทำการโหลด Python Image Library เพื่อช่วยในการจัดการกับภาพ
import requests        # ทำการโหลด request library เพื่อช่วยในการอ่านข้อมูลจาก website 
from io import BytesIO # ทำการโหลด Input Output Library เพื่อช่วยในการอ่าน binary ไฟล์

import numpy as np
import matplotlib.pyplot as plt

# กำหนด url ของภาพที่ต้องการใช้
# นักศึกษาสามารถทดลอง เปลี่ยน input ของภาพได้และสังเกตุขนาดของภาพ image.shape
url = 'https://matplotlib.org/devdocs/_images/stinkbug.png'  
#url = 'https://pixnio.com/free-images/2017/05/18/2017-05-18-10-26-37-900x580.jpg'
#url = 'https://www.mathworks.com/help/examples/matlab/win64/DisplayGrayscaleRGBIndexedOrBinaryImageExample_02.png'
response = requests.get(url)
img = Image.open(BytesIO(response.content))

print(img)  # แสดงผลข้อมูลเบื้องต้นของรูปภาพ 
print('\n')
plt.subplot(121) # สร้าง subplot ขนาด 1x2 กำหนดรูปปัจจุบัน เป็นรูปที่ 1 
plt.grid(False)  # ปิดการวาดตารางลงบนรูปภาพ
plt.imshow(img)  # ทำการเเสดงผลรูปภาพ  
plt.subplot(122) # กำหนดรูปปัจจุบัน เป็นรูปที่ 2 
plt.plot(img.histogram())  # ทำการ plot histrogram   

```

## HISTOGRAM EQUALIZATION:  
เป็นกระบวนการในการแปลง Histogram หรือลักษณะการกระจายตัวของแต่ละระดับสีให้เท่า ๆ กันทั้งช่วง   (ปรับ contrast ของภาพ)
ดูตัวอย่างตามภาพ  https://en.wikipedia.org/wiki/Histogram_equalization

![Hist](http://www.sci.utah.edu/~acoste/uou/Image/project1/images/equalization_crowd.png)
http://www.sci.utah.edu/~acoste/uou/Image/project1/Arthur_COSTE_Project_1_report.html

IDEA ในการปรับให้ระดับความเข้าของภาพนั้นกระจายอย่างทั่วถึงสามารถทำได้โดยการ
คำนวณหาค่า Cumulative distribution function (CDF)  ของ Image histrogram เริ่มต้นแล้วใช้ CDF ในการ map ระดับสีต่าง ๆ ให้การกระจายเป็นแบบ uniform 
ซึ่งเราสามารถ ใช้ function ด้านล่างนี้ในการทำ transformation  https://web.cs.wpi.edu/~emmanuel/courses/cs545/S14/slides/lecture03.pdf


จะได้สมการสำหรับการทำ image transformation ดังนี้ 

$$   f_{eq}(a) = round(  H(a)*  K-1 / MN )$$

$H(a)$  คือ CDF function 

$ a $ คือระดับสีที่ต้องการทำ

$ K จำนวนของระดับสี  = 256 $

$M \times N $ = จำนวน Pixel ในภาพขนาด M x N
